# Label Encoding

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
#load dataset
df = sns.load_dataset("titanic")
df.head()
```

```python
## one hot encoder
encode = OneHotEncoder()

encode.fit_transform(df[["embark_town"]]).toarray()
```

```python
# checking the cat names
encode.categories_
```

```python
# encode using get dummies
encode2 = pd.get_dummies(df[["embark_town", "embarked"]],drop_first=True, dtype='int')
encode2
```

```python
#loading houseprice dataset
data = pd.read_csv("homeprices.csv")
data.head()
```

```python
end1 = pd.get_dummies(data["town"], drop_first=True, dtype="int")
end1
```

```python
merged_df = pd.concat([data, end1], axis=1)
merged_df
```

```python
#drop town column
del merged_df["town"]

merged_df.head()
```

```python
#divide data into 2(features and label)

#X is our features/indepedent variables
X = merged_df.drop("price", axis=1)
X
```

```python
#dependent/label
y = merged_df["price"]
```

```python
#call the model
from sklearn.linear_model import LinearRegression
```

```python
#call/initialize the model
model = LinearRegression()

#train the model
model.fit(X,y)
```

```python
#predict the model
model.predict(X)
```

```python
merged_df["price"]
```

```python
print(round(model.score(X,y),2))
```

```python
#trying to predict unseen data

model.predict([[5000, 1, 0]])
```
