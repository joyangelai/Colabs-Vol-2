```python
# import necessary libraries
import numpy as np
import random
```

# PROJECT
Train an agent to navigate a 4x4 grid to reach a goal while avoiding traps.Think of it like teaching a robot to move through a room to find treasure without falling into holes
## Concepts applied
state: Each cell in the grid
Actions: Up, Down, Left and right
Reward: +10 for reaching the goal
-10 for falling into a trap
-1 for every other step(to encourage shorter paths)
Q-table: A table the agent uses to learn the best action for each state

```python
# create grid environment
n_states = 16
n_actions = 4  #up, down, left, right
q_table = np.zeros((n_states, n_actions))
```

```python
# define rewards
rewards = np.full(n_states, -1)
rewards[15] = 10  #Goal
rewards[5] = -10  #Trap
rewards[9] = -10  #Trap
```

```python
# parameters
alpha = 0.1  #learning rate
gamma = 0.9  #discount factor
epsilon = 0.2  #exploration rate
episodes = 500
```

```python
# Function to get next state
def get_next_state(state,action):
  row, col = divmod(state, 4)
  if action == 0 and row > 0:
    row -= 1   #up
  elif action == 1 and row < 3:
    row += 1   #down
  elif action == 2 and col > 0:
    col -= 1   # left
  elif action == 3 and col < 3:
    col += 1   #right
  return row * 4 + col
```

```python
# Q-Learning loop
for episode in range(episodes):
  state = 0    #Start at top left
  while state not in [5,9,15]:   # Until trap or goal
    if random.uniform(0,1) < epsilon:
      action = random.randint(0, 3)  # Explore
    else:
      action = np.argmax(q_table[state])  # Exploit
      next_state = get_next_state(state, action)
      reward = rewards[next_state]
      old_value = q_table[state, action]
      next_max = np.max(q_table[next_state])
```

```python
# Q- learning formula
new_value = old_value + alpha * (reward + gamma * next_max - old_value)
q_table[state, action] = new_value
state = next_state
```

```python
# Print learned Q-values
print("Trained Q-Table: ")
print(np.round(q_table, 2))
```
