```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
```

```python
df = sns.load_dataset("tips")
df.head()
```

# NORMALIZATION

```python
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df[['total_bill']])
df_scaled[:5]
```

```python
nr_tip = scaler.fit_transform(df[['tip']])
nr_tip[:5]
```

# STANDARDIZATION

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['total_bill']])
df_scaled[:5]
```
