```python
# import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, r2_score, mean_squared_error, mean_absolute_error, root_mean_squared_error
```

```python
# load dataset
df = pd.read_csv("/content/homeprices (1).csv")
df.head()
```

Observation: Data is uniform from top to bottom

```python
dataset= sns.get_dataset_names()
print(dataset, end=" ")
```

```python
df2 = sns.load_dataset("mpg")
df2.head()
```

```python
df2.columns
```

```python
df3 = df2[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
       'acceleration', 'model_year']]
df3.head()
```

```python
#  checking null values
df3.isnull().sum()
```

```python
# dropping null values
df3.dropna(inplace=True)
```

```python
# check null values again
df3.isnull().sum()
```

```python
# divide features and target
X = df3.drop('mpg', axis=1)
X.head()
y =  df3['mpg']
y.head()
```

```python
# check shape
df3.shape
```

```python
# splitting dataset to train and test
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# scaling the dataset
scale = StandardScaler()
X_train_s =scale.fit_transform(X_train)
X_test_s = scale.transform(X_test)
```

```python
# creating model & fitting
model = LinearRegression()
model.fit(X_train_s, y_train)
```

```python
# checking coeff
model.coef_
```

```python
# y intercept
model.intercept_
```

```python
# making predictions
y_pred = model.predict(X_test_s)
```

```python
# cheking accuracy
score = r2_score(y_pred, y_test)
print(round(score*100,2),"%")
```

```python
mean_squared_error(y_pred,y_test)
```

```python
mean_absolute_error(y_pred, y_test)  # --> not the best way to do it
```

```python
# doing root mean squared error
root_mean_squared_error(y_pred,y_test)
```
