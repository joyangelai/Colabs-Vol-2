```python
#import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
# Another method encode using get_dummies
# Drop-first to avoid multicollinearity / complex model / overfitting
enc1 = pd.get_dummies(df[['embark_town','embarked','sex']],drop_first=True,dtype=int)
enc1
```

```python
merged_df = pd.concat([df,enc1],axis=1)
merged_df.head()
```

```python
# dropping original columns
merged_df.drop(['embark_town','embarked','sex'], axis=1,inplace=True)
```

```python
merged_df.head()
```

```python
# loading homeprice dataset
data = pd.read_csv("/content/drive/MyDrive/homeprices.csv")
data.head()
```

```python
# encoding using get dummies
end = pd.get_dummies(data['town'], drop_first=True, dtype=int)
end.head()
```

```python
merged_df = pd.concat([data, end], axis=1)
merged_df
```

```python
# drop town column
del merged_df['town']
merged_df.head()
```

```python
# divide data into 2 (features and label)
# X is our features/independent variables
X = merged_df.drop('price', axis=1)
X
```

```python
# dependent/labe;
y = merged_df['price']
```

```python
# importing the model
from sklearn.linear_model import LinearRegression
```

```python
# call/initialise the model
model = LinearRegression()

# train the model
model.fit(X,y)
```

```python
# predict the model
model.predict(X)
```

```python
merged_df['price']
```

```python
# checking accuracy - how well the model is doing
print(round(model.score(X,y),4))
```
