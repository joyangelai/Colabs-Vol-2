```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/chinook.db")
df3 = pd.read_sql("""SELECT firstname, lastname, Title, city
                     FROM employees""", conn)
df3.head()
```

```python
# using pipes to combine firstname and lastname
conn = sqlite3.connect("/content/chinook.db")
df4 = pd.read_sql("""SELECT firstname || ' ' || lastname as FullName, Title, city
                     FROM employees""", conn)
df4.head()
```

```python
# select names starting with ste
conn = sqlite3.connect("/content/chinook.db")
dfa= pd.read_sql("""SELECT firstname, Title, city
                     FROM employees
                     WHERE firstname LIKE 'Ste%'
                      """, conn)
dfa.head()
```

```python
# select names ending with ve
conn = sqlite3.connect("/content/chinook.db")
dfa= pd.read_sql("""SELECT firstname, Title, city
                     FROM employees
                     WHERE firstname LIKE '%ve'
                      """, conn)
dfa.head()
```

```python
# selecting from calgary city
conn = sqlite3.connect("/content/chinook.db")
dfb= pd.read_sql("""SELECT firstname, Title, city
                     FROM employees
                     WHERE city LIKE '%Calgary%'
                      """, conn)
dfb.head()
```

```python
# selecting from calgary city using in operator
conn = sqlite3.connect("/content/chinook.db")
dfc= pd.read_sql("""SELECT firstname, Title, city
                     FROM employees
                     WHERE city IN ('Calgary', 'Lethbridge')
                      """, conn)
dfc
```

```python
# checking unique cities in ascending order(ASC) and decscending order (DESC)
dfz= pd.read_sql("""SELECT DISTINCT (city)
                     FROM employees
                     ORDER BY city ASC
                      """, conn)
dfz.head()
```

```python
# using group by
dfy= pd.read_sql("""SELECT BillingCity, COUNT(CustomerId) as total_count
                     FROM invoices
                     GROUP BY BillingCity
                     HAVING COUNT(Customerid) > 7
                     ORDER BY Billingcity ASC
                      """, conn)
dfy
```
